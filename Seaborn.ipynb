# Distribution Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
# sns.set(style='whitegrid')
sns.displot(data=df,x='salary',kde=True,rug=True)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='age')
```

```python
sns.displot(data=sample_ages,x='age',rug=True,bins=30,kde=True)
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100],bw_adjust=0.1)
```

# Categorical Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='level of education')
```

```python
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci='sd',hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
```
