# Scatter Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='Dark2')
plt.show()
```

# Categorical Plots
- For single category

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.countplot(data=df, x='division')
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.countplot(data=df, x='level of education', hue='division', palette='Set2')
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=150)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

# Distribution Plots

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4), dpi=150)
sns.rugplot(x='salary', data=df, height=0.5)
plt.show()
```

```python
sns.set(style='white')
sns.displot(data=df, x='salary', bins=20,kde=True, rug=True)
plt.show()
```

```python
sns.kdeplot(data=df, x='salary')
plt.show()
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages, x='age')
plt.show()
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30, kde=True)
plt.show()
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.6, shade=True)
plt.show()
```

# Box plots and Swarmplots

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score',y='gender', size=2, hue='test preparation course', dodge=True)
plt.legend(bbox_to_anchor=(1.35, 0.5))
plt.show()
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.boxenplot(data=df, x='math score', y='test preparation course', hue='gender')
plt.legend(bbox_to_anchor=(1.35, 0.5))
plt.show()
```

# Grid Plots

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', row='lunch')
plt.show()
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch', row='test preparation course')
plt.show()
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', row='race/ethnicity', col='lunch')
plt.show()
```

```python
g = sns.PairGrid(data=df, hue='gender')

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.kdeplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend()
```
