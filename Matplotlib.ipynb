```python
fig = plt.figure()
# Large Axis

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# Small Axis
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In');
```

```python
fig = plt.figure(figsize=(10,3),dpi=100)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
fig.savefig('new_figure.png',bbox_inches='tight')
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y=2*x
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)
axes[0].plot(x,y)
axes[1].plot(a,b)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label='X vs X^2')
ax.legend()
```
